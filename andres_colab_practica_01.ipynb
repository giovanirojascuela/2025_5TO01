ejercicio 03 data frame:

```python
import pandas as pd
#creando un dataframe de un  data set
df=pd.DataFrame(data={"pais":["mexico","argentina","españa","colombia"],
                      "poblacion":[127212000,45167000,47099000,48922000]})
```

```python
df
```

```python
# ordenar columnas por variables
df.sort_values(["pais"])
```

```python
#agregando columna o variable
df["Superficie"]=[1964375,27884000,505944.1142748]
df
```

```python
#agregar loso un valor a las duplas de una variable
df["Deporte"]="futbol"
df
```

```python
#eliminar columna o variable
df=df.drop(["Deporte"],axis=1)
df
```

```python
#eliminar varias variables
df2=df
##eliminacion multiple
df2.drop(["pais"])
df
```

```python
#numero de tuplas
Cantidad_filas=len(df)
Cantidad_filas
```

```python
# agregacion de filas
df.loc[Cantidad_filas]=["venezuela",0,9116445]
df
```

```python
#cambio de valor
df.at[4,"poblacion"]=[48108600]
df
```

```python
# eliminacion por celdas (valor de una tupla)
df2=df
df2.drop(1,axis=0)
```
