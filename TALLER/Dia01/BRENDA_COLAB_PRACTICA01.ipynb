Ejercicio03-DATAFRAME

```python
import pandas as pd

# Creando un DataFrame desde un dataset
# Asegúrate de que las listas de cada columna tengan la misma longitud
df=pd.DataFrame(data={"Pais":["Mexico","Argentina","colombia","China","Brazil"],
                      "poblacion":[124585100,54872100,454742000,46872100, 214326223]}) # Agregado un valor para Brazil
```

```python
#mostrar dataframe
df
```

```python
#ordenar columnas por variable
df.sort_values(by="poblacion",ascending=False)
```

```python
#agregando columna o variable
# Asegúrate de que la lista de valores tenga la misma longitud que el número de filas en el DataFrame (5 en este caso)
df["Superficie"]=[1964375, 2788470, 505945, 9596961, 8515767] # Agregado un valor para Brazil y corregido algunos valores para que sean mas realistas
df
```

```python
#agregar solo un valor a las tuplas  de una variable
df["Deporte"]="Futbol"
df
```

```python
#eliminar columna o variable
df=df.drop(["Deporte"],axis=1)
df
```

```python
df2=df
df2
#eliminacion multiple de variables.
df2=df2.drop(["Pais","Superficie"],axis=1)
df2
```

```python
#numero de tuplas
Cantidad_filas=len(df)
Cantidad_filas
```

```python
#agregacion de filas
df.loc[Cantidad_filas] = ["Nuevo Pais", 0, 0] # Reemplaza "Nuevo Pais", 0, 0 con los valores reales que deseas agregar
df
```

```python
#cambio de valor de una variable en una tupla.
df.at[4,"poblacion"]=[485548000]
df
```

```python
#eliminacion por celda(valor de una variable en una tupla)
df2=df
df2.drop(1,axis=0)
```
