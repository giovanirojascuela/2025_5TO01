#CREA Y ENTRENA UNA RED NUERONAL CON PYTHON TENSORFLOW

**1.- Cargamos y exploramos datos Fashion MINIST**

```python

from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
import numpy as np
#cargar el datasets fashion minist
(X_train,Y_train),(X_test,Y_test)=fashion_mnist.load_data()
#exploramos dataset
labels=["T.shirt/top","Trouser","Pulover","Dres","Coat","Sandal",
        "Shirt","Sneaker","Bag","Ankle boot"]
plt.figure(figsize=(14,8))
for i in range(20):
  plt.subplot(4,5,i+1)
  plt.imshow(X_train[i],cmap="binary")
  plt.title(labels[Y_train[i]])
  plt.axis("off")
plt.show()
```

**2.- Normalizar datos**

```python
X_train,X_test=X_train/255.0,X_test/255.0
```

**3.- Definicion de arquitectura de la red nueronal**

```python

from tensorflow import keras
model=keras.Sequential([
    keras.layers.Flatten(input_shape=[28,28]),
    keras.layers.Dense(128,activation="relu"),
    keras.layers.Dense(10,activation="softmax")
])
```

```python
#Compilar modelo
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

**4.- Entrenar Modelo**

```python

model.fit(X_train,Y_train,epochs=10,validation_data=(X_test,Y_test))
```

```python
#Evaluar modelo
test_loss,test_acc=model.evaluate(X_test,Y_test)
print("\nPresicion en el conjunto de prueba",test_acc)
```
