#CREA Y ENTRENA UNA RED NEURONAL CON PYTHON Y TENSORFLOW


**CARGAMOS Y EXPLORAMOS DATOS FASHION MINIST**

```python
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
import numpy as np
#cargar el dataset fashion Mnist
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
#Exploramos dataset
labels=["T-shirt/top","Trouser","Pullover","Dres","Coat",
        "Sandal","Shirt","Sneaker","Bag","Ankle boot"]
plt.figure(figsize=(14,8))
for i in range(20):
  plt.subplot(4,5,i+1)
  plt.imshow(X_train[i],cmap="binary")
  plt.xlabel(labels[y_train[i]])
  plt.axis('off')
plt.show()
```

```python
X_train,X_test=X_train/255.0,X_test/255.0
```

**3. definicion de arquitecto de la red neuronal**

```python
from tensorflow import keras
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128,activation="relu"),
    keras.layers.Dense(10,activation="softmax")
])
```

```python
#compilar modelo
model.compile(
              optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"]
              )
```

**4.entrenar modelo**

```python
model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test))
```

```python
test_loss,test_acc=model.evaluate(X_test,y_test)
print("\nPresicion en el conjunto de prueba",test_acc)
```
