**Ejercicio 01-Regrasion lineal**

```python
#importacion de librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#creando database
temperatura=np.array([15,16,18,20,21,23,25,27,30,32])
ventas=np.array([500,520,560,580,600,640,680,700,760,800])
print(temperatura)
print(ventas)
```

```python
#Dividimos los datos en conjuntos de entrenamiento y pruebas
X_train, X_test, Y_train, Y_test=train_test_split(temperatura.reshape(-1,1),
                                                  ventas,test_size=0.2,
                                                  random_state=42)
print("entrenamiento X------------------------")
print(X_train)
print("Prueba X--------------------")
print(X_test)
print("entrenamiento Y-------------------")
print(Y_train)
print("prueba Y--------------------")
print(Y_test)
```

```python
#creamos el modelo y lo entrenamos
modelo=LinearRegression()
modelo.fit(X_train,Y_train)
```

```python
# Predicción y gráficos
plt.scatter(temperatura, ventas, color="blue", label="Datos reales")
plt.plot(temperatura, modelo.predict(temperatura.reshape(-1, 1)),
         color="red", label="Línea de regresión")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Ventas")
plt.legend()
plt.show()


```

```python
r_sq=modelo.score(X_test,Y_test)
print("coeficiente de determinacion",r_sq)
```
