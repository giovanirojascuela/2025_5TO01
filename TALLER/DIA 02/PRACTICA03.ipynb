# PRACTICA 03 SKLEARN-tensorflow-numpy


** Ejercicio 01-Regrecion Lineal

```python
#Importacion de libreria
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#creando dataset
temperatura=np.array([15,16,18,20,21,23,25,27,30,32])
ventas=np.array([500,520,560,580,600,640,680,700,760,800])
print(temperatura)
print(ventas)

```

```python
#Dividimos los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test=train_test_split(temperatura.reshape(-1,1),
                                                  ventas,test_size=0.2,
                                                  random_state=42)
print("Entrenamiento x-----------------")
print(X_train)
print("Prueba x-----------------")
print(X_test)
print("Entrenamiento y----------------")
print(y_train)
print("Prueba Y----------------")
print(y_test)

```

```python
#Creamos el modelo y lo entremanos
modelo=LinearRegression()
modelo.fit(X_train,y_train)
```

```python
#Prediccion y grafico
plt.scatter(temperatura,ventas,color="blue",label="Datos Reales")
plt.plot(temperatura,modelo.predict(temperatura.reshape(-1,1)),
          color="red",label="Linea de Regresion")
plt.xlabel("Temperatura (°c)")
plt.xlabel("ventas")
plt.legend()
plt.show()
```

```python
#Evaluacion de Resultados
r_sq=modelo.score(X_test,y_test)
print("coeficiente de determinacion",r_sq)
```
