#Principal Component Analysis (PCA): Reducci贸n de dimensionalidad para visualizar datos de c谩ncer.

```python
#  PASO 1: Importar librer铆as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Configurar estilo visual
sns.set(style="whitegrid")
```

```python
#  PASO 2: Cargar el dataset original de Kaggle (subido como 'data.csv')
df = pd.read_csv('data.csv')

# Mostrar las primeras filas
df.head()
```

```python
#  PASO 3: Explorar el dataset
print("Dimensiones del dataset:", df.shape)
print("\nColumnas del dataset:")
print(df.columns)

# Eliminar columnas irrelevantes: 'id' y 'Unnamed: 32'
df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)

# Contar valores por clase (diagn贸stico)
sns.countplot(data=df, x='diagnosis', palette='Set2')
plt.title('Distribuci贸n de Diagn贸stico (B = Benigno, M = Maligno)')
plt.show()

# Ver estad铆sticas descriptivas
df.describe()
```

```python
#  PASO 4: An谩lisis de correlaci贸n
plt.figure(figsize=(12, 8))
sns.heatmap(df.iloc[:, 1:].corr(), cmap='coolwarm', linewidths=0.5)
plt.title("Mapa de Calor de Correlaciones entre Caracter铆sticas")
plt.show()
```

```python
# 锔 PASO 5: Preparar los datos para PCA

# Convertir diagn贸stico a valores num茅ricos: M = 1, B = 0
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})

# Separar variables independientes y dependiente
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

# Estandarizar las variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#  PASO 6: Aplicar PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Crear nuevo DataFrame para visualizaci贸n
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['diagnosis'] = y
```

```python
#  PASO 7: Visualizaci贸n 2D de PCA
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='diagnosis', palette='Set1')
plt.title('Proyecci贸n PCA (2 Componentes) - Diagn贸stico de C谩ncer')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Diagn贸stico', labels=['Benigno (0)', 'Maligno (1)'])
plt.grid(True)
plt.show()
```

```python
#  PASO 8: Varianza explicada por componentes
print("Varianza explicada por cada componente:")
for i, var in enumerate(pca.explained_variance_ratio_):
    print(f"PC{i+1}: {var:.4f}")

print(f"\n Varianza total explicada por los 2 componentes: {sum(pca.explained_variance_ratio_):.2%}")
```

```python
# П PASO 9: Visualizar distribuci贸n de diagn贸stico en componentes individuales
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(data=pca_df, x='diagnosis', y='PC1', ax=axes[0], palette='Set2')
axes[0].set_title('Distribuci贸n de PC1 por Diagn贸stico')

sns.boxplot(data=pca_df, x='diagnosis', y='PC2', ax=axes[1], palette='Set3')
axes[1].set_title('Distribuci贸n de PC2 por Diagn贸stico')

plt.tight_layout()
plt.show()
```
