```python
import pandas as pd
# ceando undatagframe de un dataset
df = pd.DataFrame(data={
    "Pais": ["Brazil", "Argentina", "Portugal", "Peru"],
    "Poblacion": [1234323, 134234543, 344354645, 12423423]})
df ["Superficie"]=[1234323, 134234543, 344354645, 12423423]
Cantidad_filas=len(df)
df.loc[Cantidad_filas]=["Venezuela",346456745743534,9116445]
df.at[4,"Poblacion"]=[4810600]
df
```

```python
#Preparacion de dataframe
#agregando variable radio de habitante por superficie
df["habitxkm2"]=(df["Poblacion"]/df["Superficie"]).astype(int)
df.sort_values(["habitxkm2"])
```

```python
#Importar la libreria matplotib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df[["Poblacion", "Superficie"]].plot.hist(bins=20, alpha=0.5)
```

```python
# Grafico de barras
df.set_index("Pais")["Poblacion"].plot(kind="bar")
```

```python
#Grafco de area
df.set_index("Pais")["habitxkm2"].plot(kind="area")
```

```python
#grafico de pastel
df.set_index("Pais")["Superficie"].plot(kind="pie")
```
