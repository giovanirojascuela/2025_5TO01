# PRCATICA 03 SKLEARN-TENSORFLOW - NUMPY



**EJERCICIO 01 REGRESION LINEAL**



```python
#Importacion de librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#Creando dataset
temperatura=np.array([15,16,18,20,21,23,25,27,30,32])
ventas=np.array([500,520,560,580,600,640,680,700,760,800])

print(temperatura)
print(ventas)
```

```python
#Dividimos los datos en conjuntos de entrenamiento y prueba
X_train,X_test,Y_train,Y_test=train_test_split(temperatura.reshape(-1,1),
                                               ventas,test_size=0.2,
                                               random_state=42)

print("Entrenamientp X---------------------------")
print(X_train)
print("Prueba X--------------------------------------")
print(X_test)
print("Entrenamientp Y---------------------------")
print(Y_train)
print("Prueba Y--------------------------------------")
print(Y_test)
```

```python
#creando el modelo y entrenamos
modelo=LinearRegression()
modelo.fit(X_train,Y_train)
```

```python
#prediccion y grafico
plt.scatter(temperatura,ventas,color="blue",label="Datos Reales")
plt.plot(temperatura,modelo.predict(temperatura.reshape(-1,1)),
         color="red",label="Linea de regresion")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Ventas")
plt.legend()
plt.show()
```

```python
#evaluacion de resultados
r_sq=modelo.score(X_test,Y_test)
print("Coeficiente de determinacion",r_sq)
```
