```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#Creando dataset
X=np.array([10,11,13,15,13,12,10,11,15,10])
Y=np.array([15,16,18,19,17,16,16,17,20,16])
xtr,xts,yt,yts=train_test_split(X.reshape(-1,1),
                                                Y,test_size=0.2,
                                                random_state=42)
print(X)
print(Y
      )
```

```python
print ("Entrenamineto x ------------")
print(xtr)
print ("Prueba x ------------")
print(xts)
print ("Entrenamineto y ------------")
print(yt)
print ("Prueba y ------------")
print(yts)
```

```python
# prompt: crea el modelo y ewntrenalo

# Crea el modelo de regresi√≥n lineal
model = LinearRegression()

# Entrena el modelo
model.fit(xtr, yt)

```

```python
# prompt: arregla este codigo, me dice que modelo no esta definido: plt.scatter(temp,v,color="blue",label="Datos reales")
# plt.plot(temp,modelo.predict(temp.reshape(-1,1)),
#          color="red",label="Linea de regresion")
# plt.xlabel("Temperatura")
# plt.ylabel("Ventas")
# plt.legend()
# plt.show()

import matplotlib.pyplot as plt
plt.scatter(X,Y,color="blue",label="Datos reales")
plt.plot(X,model.predict(X.reshape(-1,1)),
         color="red",label="Linea de regresion")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
#Evaluacion de Resultados
r_sq=model.score(xts,yts)
print("Coeficiente de determinacion: ",r_sq)
```
