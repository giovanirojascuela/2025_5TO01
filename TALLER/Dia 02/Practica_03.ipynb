## PRACTICA 03 sklearn-tensorflow-numpy

**Ejercicio 01- Regrecion lineal**

```python
#importacion de librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#Creando dataset
temperatura=np.array([15,16,18,20,21,23,25,27,30,32])
ventas=np.array([500,520,560,580,600,640,680,700,760,800])
print(temperatura)
print(ventas)
```

```python
# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, Y_train, Y_test=train_test_split(temperatura.reshape(-1,1),
                                                  ventas,test_size=0.2,
                                                  random_state=42)
print("Entrenamiento x-------------------")
print(X_train)
print("Prueba x-------------------------")
print(X_test)
print("Entrenamiento y-------------------------")
print(Y_train)
print("Prueba y-------------------------")
print(Y_test)
```

```python
#Creamos el modelo y lo entrenamos
modelo=LinearRegression()
modelo.fit(X_train,Y_train)
```

```python
#prediccion yGrafico
plt.scatter(temperatura,ventas,color="blue",label="Datos Reales")
plt.plot(temperatura,modelo.predict(temperatura.reshape(-1,1)),
         color="red",label="Regresion Lineal")
plt.xlabel("Temperatura (°c)")
plt.ylabel("Ventas")
plt.show()
```

```python
#evaluacion de resultados
r_sq=modelo.score(X_train,Y_train)
print("Coeficiente de determinacion: ",r_sq)
```
