```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#Creando dataset
temp=np.array([15,16,17,20,21,23,25,27,30,32])
v=np.array([500,520,560,580,600,640,680,700,760,800])
xtr,xts,yt,yts=train_test_split(temp.reshape(-1,1),
                                                v,test_size=0.2,
                                                random_state=42)
print(temp)
print(v
      )
```

```python
print ("Entrenamineto x ------------")
print(xtr)
print ("Prueba x ------------")
print(xts)
print ("Entrenamineto y ------------")
print(yt)
print ("Prueba y ------------")
print(yts)
```

```python
# prompt: crea el modelo y ewntrenalo

# Crea el modelo de regresión lineal
model = LinearRegression()

# Entrena el modelo
model.fit(xtr, yt)

```
