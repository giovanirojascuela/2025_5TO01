```python
#Importacion de librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#Creando dataset
X=np.array([10,11,13,15,13,12,10,11,15,10])
Y=np.array([15,16,18,19,17,16,16,17,20,16])
print(X)
print(Y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X.reshape(-1,1),Y,test_size=0.2,random_state=42)
#Creamos el modelo y lo entrenamos
modelo=LinearRegression()
modelo.fit(X_train,y_train)
```

```python
#Realizamos predicciones en el conjunto de prueba y graficamos
plt.scatter(X,Y,color="blue",label="Datos Reales")
plt.plot(X,modelo.predict(X.reshape(-1,1)),color="red",label="Linea de Regresion")
plt.xlabel("Datos X")
plt.ylabel("Datos Y")
plt.title("Regresion Lineal")
plt.legend()
plt.show()
```

```python
r_sq=modelo.score(X_test,y_test)
print("Coeficiente de determinacion ",r_sq)
```
