# PRACTICA 03 SKLEARN-TENSORFLOW

### Ejercicio 01-Regresion lineal

```python
#importacion de librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#creando dataset
temperatura=np.array([15,16,18,20,21,23,25,27,30,32])
ventas=np.array([500,520,560,580,600,640,680,700,760,800])
print(temperatura)
print(ventas)
```

```python
#Dividimos datos en conjuntos de entrenamiento y prueba
X_train,X_test, Y_train, Y_test=train_test_split(temperatura.reshape(-1,1),ventas,test_size=0.2,random_state=42)
print("Entrenamiento--X----------------")
print(X_train)
print("Entrenamiento--Y----------------")
print(Y_train)
print("Prueba--X----------------")
print(X_test)
print("Prueba--Y----------------")
print(Y_test)
```

```python
#Creando el ,modelo y lo entrenamos
modelo=LinearRegression()
modelo.fit(X_train,Y_train)
```

```python
#Realizamos predicciones en el conjunto de prueba y graficamos
plt.scatter(temperatura,ventas,color="blue",label="Datos Reales")
plt.plot(temperatura,modelo.predict(temperatura.reshape(-1,1)),color="red",label="Linear de Regresion")
plt.xlabel("Temperatura (C°)")
plt.ylabel("Ventas")
plt.legend()
plt.show()
```

```python
#Evaluacion de resultados
r_sq=modelo.score(X_test,Y_test)
print("Coeficiente de determinacion (R^2):",r_sq)
```
