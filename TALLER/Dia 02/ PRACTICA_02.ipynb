```python
import pandas as pd
df=pd.DataFrame(data={"Pais":["Mexico", "Agentina","España","Colombia"],
                      "Poblacion":[1964375,134344543,344354645,12423423]})
df["Superficie "]=[1234323,134344543,344354645,12423423]
Cantidad_filas=len(df)
df.loc[Cantidad_filas]=["Venezuela",0,9116445]
df.at[4,"Poblacion"]=[4810600]
df
```

```python
#Preparacion de dataframe
#Agregando variable radio de habitante por superficie
df["habitxkm2"]=(df["Poblacion"]/df["Superficie "]).astype(int)
df.sort_values("habitxkm2")
```

```python
#Importar libreria matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df[["Poblacion", "Superficie"]].hist(bins=20, alpha=0.5)
```

```python
# grafico de barras
df.set_index("Pais")["Poblacion"].plot(kind="bar")
```

```python
#Grafico de area
df.set_index("Pais")["habitxkm2"].plot(kind="area")
```

```python
#Grafico de pastel
df.set_index("Pais")["Superficie"].plot(kind="pie")
```
