## PRACTICA 03 sklearn-tensorflow

**Ejercicio 01 - regresion lineal**

```python
#importacion de librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#creando dataset
temperatura=np.array([15,16,18,20,21,23,25,27,30,32])
ventas=np.array([500,520,560,580,600,40,680,700,760,800])
print(temperatura)
print(ventas)
```

```python
##dividimos los datos en conjunto de entrenamiento y prueba
X_train, X_test, Y_train, Y_test=train_test_split(temperatura.reshape(-1,1),ventas,test_size=0.2,random_state=42)
print("EntrenamientoX------------")
print(X_train)
print("E-X-----------------------")
print(X_test)
print("E-X-----------------------")
print(Y_train)
print("E-Y-----------------------")
print(Y_test)
print("E-Y-----------------------")
```

```python
#Creamos el modelo y lo entrenamos
modelo=LinearRegression()
modelo.fit(X_train,Y_train)
```

```python
plt.scatter(temperatura,ventas,color="blue",label="Datos Reales")
plt.plot(temperatura,modelo.predict(temperatura.reshape(-1,1)),
         color="red",label="Lina de regresion")
plt.xlabel("Temperatura(°C)")
plt.ylabel("ventas")
plt.legend()
plt.show()
```

```python
#EVALUACION DE RESULTADOS
r_sq=modelo.score(X_test,Y_test)
print("Coeficiente de determinacion",r_sq)
```
