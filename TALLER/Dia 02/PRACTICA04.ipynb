```python
#importacion de librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#CREANDO UN DATASET
temperatura=np.array([10, 11, 13, 15, 13, 12, 10, 11, 15, 10])
# Removed the last element (800) from the ventas array to match the length of temperatura
ventas=np.array([15, 16, 18, 19, 17, 16, 16, 17, 20, 16])
print(temperatura)
print (ventas)
```

```python
# DIVIDIMOS LOS DATOS EN COJUNTO DE ENTRENAMIENTO  Y PRUEBAS
X_train,X_test,y_train,y_test=train_test_split(temperatura.reshape(-1,1),
                                               ventas,test_size=0.2,
                                               random_state=42)
print("Entrenamiento x-----------------------")
print(X_train)
print("prueba x--------------------------------")
print(X_test)
print("Entrenamiento y-----------------------")
print(y_train)
print("Prueba Y-------------------------------")
print(y_test)
```

```python
#CREAMOS EL MODELO Y LO ENTRENAMOS
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
#PREDICCION Y GRAFICOS
plt.scatter(temperatura,ventas,color="blue",label="Datos Reales")
plt.plot(temperatura,model.predict(temperatura.reshape(-1,1)),
         color="red",label="Linea de Regresion")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Ventas")
plt.title("Regresion Lineal")
plt.legend()
plt.show()
```

```python
#EVALUACION  DE RESULTADO
r_sq=model.score(X_test,y_test)
print("Coeficiente de Determinacion",r_sq)
```
