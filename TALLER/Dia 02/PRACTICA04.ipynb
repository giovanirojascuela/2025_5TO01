# PRACTICA 03 SKLEARN-tensorflow-numpy


** Ejercicio 01-Regrecion Lineal

```python
#Importacion de libreria
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#creando dataset
x=np.array([10,11,13,15,13,12,10,11,15,10])
y=np.array([15,16,18,19,17,16,16,17,20,16])
print(x)
print(y)

```

```python
#Dividimos los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test=train_test_split(x.reshape(-1,1),
                                                  y,test_size=0.2,
                                                  random_state=42)
print("Entrenamiento x-----------------")
print(X_train)
print("Prueba x-----------------")
print(X_test)
print("Entrenamiento y----------------")
print(y_train)
print("Prueba Y----------------")
print(y_test)

```

```python
#Creamos el modelo y lo entremanos
modelo=LinearRegression()
modelo.fit(X_train,y_train)
```

```python
#Prediccion y grafico
plt.scatter(x,y,color="blue",label="Datos Reales")
plt.plot(x,modelo.predict(x.reshape(-1,1)),
          color="red",label="Linea de Regresion")
plt.xlabel("x (°c)")
plt.xlabel("y")
plt.legend()
plt.show()
```

```python
#Evaluacion de Resultados
r_sq=modelo.score(X_test,y_test)
print("coeficiente de determinacion",r_sq)
```
