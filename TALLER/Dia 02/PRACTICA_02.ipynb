```python
import pandas as pd
#CREANDO UN DATAFRAME DE UN DATASET
import matplotlib.pyplot as plt
df=pd.DataFrame(data={"Pais":["Mexico","Argentina","España","Colombia"],
                      "Poblacion":[127212000,45167000,47099000,48922000]})
df["superficie"]=[1964555,2457822,585635,1554844]
Cantidad_filas=len(df)
df.loc[Cantidad_filas]=["Venezuela",0,9116445]
df.at[4,"Poblacion"]=[48108600]
(df)
```

```python

#PREPARACION DE DATAFRAME
#agregamos variables radio de habitante por superficie
df["habitxkm2"]=(df["Poblacion"]/df["superficie"]).astype(int)
df.sort_values(["habitxkm2"])
```

```python
#importar libreria matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df[["Poblacion","superficie"]].plot.hist(bins=20,alpha=0.5)
```

```python
#grafico de barras
df.set_index("Pais")["Poblacion"].plot(kind="bar")
```

```python
#grafico de area
df.set_index("Pais")["habitxkm2"].plot(kind="area")
```

```python
#GRAFICO DE PASTEL
df.set_index("Pais")["superficie"].plot(kind="pie")
```
