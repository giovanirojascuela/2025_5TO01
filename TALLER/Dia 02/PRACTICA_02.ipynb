** PRACTICA 02 GRAFICO CON PANDAS


```python
import pandas as pd

# Creando un dataframe con datos reales aproximados
df = pd.DataFrame(data={
    "Pais": ["Mexico", "Argentina", "España", "Colombia", "Venezuela"],
    "Poblacion": [126014024, 45195777, 47351567, 50882884, 28435943],  # datos de población aproximada 2023
    "Superficie": [1964375, 2780400, 505990, 1141748, 916445]  # en km², superficie real aproximada
})

# No necesitamos añadir una fila duplicada ni modificar la población si ya es correcta
df
```

```python
#Preparacion de dataframe
df["habitxKm2"]=(df["Poblacion"]/df["Superficie"]).astype(int)
df.sort_values(["habitxKm2"])
```

```python
#Importar MATPLOTLIB
import matplotlib as plt
%matplotlib inline
```

```python
df[["Poblacion","Superficie"]].plot.hist(bins=20,alpha=0.5)
```

```python
df.set_index("Pais")["Poblacion"].plot(kind="bar")
```

```python
#GRAFICO DE AREA
df.set_index("Pais")["habitxKm2"].plot(kind="area")
```

```python
#GRAFICO DE PASTEL
df.set_index("Pais")["Superficie"].plot(kind="pie")
```
