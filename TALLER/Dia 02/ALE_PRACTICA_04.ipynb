```python
#Importacion de librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#Creando dataset
X=np.array([10,11,13,15,13,12,10,11,15,10])
Y=np.array([15,16,18,19,17,16,16,17,20,16])

print(X)
print(Y)
```

```python
#Dividimos los datos en conjuntos de entrenamiento y prueba
X_train,X_test,Y_train,Y_test=train_test_split(X.reshape(-1,1),
                                               Y,test_size=0.2,
                                               random_state=42)

print("Entrenamientp X---------------------------")
print(X_train)
print("Prueba X--------------------------------------")
print(X_test)
print("Entrenamientp Y---------------------------")
print(Y_train)
print("Prueba Y--------------------------------------")
print(Y_test)
```

```python
#creando el modelo y entrenamos
modelo=LinearRegression()
modelo.fit(X_train,Y_train)
```

```python
#prediccion y grafico
plt.scatter(X,Y,color="blue",label="Datos Reales")
plt.plot(X,modelo.predict(X.reshape(-1,1)),
         color="red",label="Linea de regresion")
plt.xlabel("X (°C)")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
#evaluacion de resultados
r_sq=modelo.score(X_test,Y_test)
print("Coeficiente de determinacion",r_sq)
```
