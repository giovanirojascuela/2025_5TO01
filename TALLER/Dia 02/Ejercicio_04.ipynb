```python
# Importacion de librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
# Creando dataset
x=np.array([10,11,13,15,13,12,10,11,15,10])
y=np.array([15,16,18,19,17,16,16,17,20,16])
print(x)
print(y)
```

```python
# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, Y_train, Y_test=train_test_split(x.reshape(-1,1),
                                                y,test_size=0.2,
                                                random_state=42)

print("Entrenamiento--------------")
print(X_train)
print("Prueba--------------")
print(X_test)
print("Entrenamiento--------------")
print(Y_train)
print("Prueba--------------")
print(Y_test)
```

```python
# Creamos el modelo y lo entrenamos
modelo=LinearRegression()
modelo.fit(X_train,Y_train)
```

```python
# Prediccion y Grafico
plt.scatter(x, y, color="blue", label="Datos Reales")
plt.plot(x, modelo.predict(x.reshape(-1, 1)), color="red", label="Línea de Regresión")
plt.xlabel("x ")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# Evaluacion de resultado
r_sq=modelo.score(X_test,Y_test)
print("Coeficiente de Determinacion (R^2):",r_sq)
```
