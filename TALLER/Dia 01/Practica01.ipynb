Ejercicio 03-DAtaFrame


```python
import pandas as pd
#creando un dataframe de un dataset
df=pd.DataFrame(data={"Pais":["Mexico","Argentina","España","Colombia"],
                      "Poblacion":[127212000,45167000,47099000,48922000]})
```

```python
#mostramos DataFrame
df
```

```python
#ordenar columnas por variables
df.sort_values(['Pais'])
```

```python
#agregando columna o variable
df["Superficie"]=[1964374,32492834,5034823,39128423]
df
```

```python
#agregar solo un valor a las duplas de una variable
df["Deporte"]="Futbol"
df
```

```python
#eliminar columna o variable
df=df.drop(["Deporte"],axis=1)
df
```

```python
#Eliminacion multiple de variables
df2=df

df2.drop(["Pais","Superficie"],axis=1)
```

```python
#numero de tuplas
Cantidad_filas=len(df)
Cantidad_filas
```

```python
#agregacion de filas
df.loc[Cantidad_filas]=["Venezuela",0,91164445]
df
```

```python
#cambiar
df.at[4,"Poblacion"]=[48108600]
df
```

```python
#Eliminacion por celda(valor de una variable en una dupla)
df2=df
df.drop(1,axis=0)
```

Ejercicio 04 - filtrado de DataFrames

```python
#operacipones de comparacion y logicos
#que pais tiene mas de 46 millones
mas_de_46=df["Poblacion"]>48000000
df[mas_de_46]
```

```python
#filtrado  de comparacion de logicos
df[(df["Poblacion"]>48000000)&(df["Superficie"]>60000000)]
```

```python
#filtrado por valor de variable
por_nombre=df["Pais"]=="Mexico"
df[por_nombre]
```

```python
#filtrado por tamaño de valor de variable
nombre_largo=df["Pais"].str.len()>6
df[nombre_largo]
```

```python
#filtrado por indice
por_nombre.index.to_list()[0]
```
