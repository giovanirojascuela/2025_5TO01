**Ejercici0 03 - DataFrame**

```python
import pandas as pd
#creando un dataframe de un dataset
df=pd.DataFrame(data={"Pais":["Mexico","Argentina","España","Colombia"],
                      "Poblacion":[127212000,45167000,47099000,48922000]})
```

```python
#Mostrando DataFrame
df
```

```python
#Ordenar columnas por variable
df.sort_values(["Pais"])
```

```python
#Agregando columna o variable
df["Superficie"]=[1964375,2788400,505944,1142748]
df
```

```python
#Agregar solo unvalor a las tuplas de una variable
df["Deporte"]="Futbol"
df
```

```python
#Elimar columna o variable
df=df.drop(["Deporte"],axis=1)
df
```

```python
df2=df
df2
#Eliminacion multipe de variables
df2.drop(["Pais","Superficie"],axis=1)
```

```python
#Numero de tuplas
Cantidad_filas=len(df)
Cantidad_filas
```

```python
#Agregacion de filas
df.loc[Cantidad_filas]=["venezuela",0,91164445]
df
```

```python
#Cambio de valor de una variable en una tupla
df.at[4,"Poblacion"]=[48108600]
df
```

```python
#Eliminacion por celda(valor de una variable en una tupla)
df2=df
df2.drop(1,axis=0)
```

**Ejercicio 04 - filtrado de DataFrame**

```python
#Operadores de comparacion y logicos
#Que pais tiene mas de 46 millones
mas_de_46=df[df["Poblacion"]>46000000]
mas_de_46
```

```python
#Filtrando con operadores logicos
df[(df["Poblacion"]>48000000)&(df["Superficie"]<46000000)]
```

```python
#Filtrando por valores de variable
por_nombre=df[df["Pais"]=="Colombia"]
por_nombre
```

```python
#filtrando por tamaño de valor de variable
nombre_largo=df[df["Pais"].str.len()>6]
nombre_largo
```

```python
#Filtrado por indice
por_nombre.index.to_list()[0]
```
