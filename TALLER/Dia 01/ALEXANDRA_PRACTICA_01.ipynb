** Ejercicio 03 - dataframe **


```python
import pandas as pd
# creando un dataframe de un dataset
df=pd.DataFrame(data={"Pais":["Mexico","Argentina","España","Colombia"],
                      "Poblacion":[127212000,45167000,47099000,48922000]})
```

```python
#Mostrando dataframe
df
```

```python
#Ordenar columnas por variable
df.sort_values(["Pais"])
```

```python
#Agregando columna o variable
df["Superficie"]=[1964375,2788400,505944,1142748]
df
```

```python
#agregar solo un valor a la tuplas de una variable
df["Deporte"]="Futbol"
df
```

```python
#Elimar columna o variable
df=df.drop(["Deporte"],axis=1)
df
```

```python
df2=df
df2
#Eliminacion multiple de variables
df2.drop(["Pais","Superficie"],axis=1)
```

```python
#Numero de tuplas
cantidad_filas=len(df)
cantidad_filas
```

```python
#Agregacion de filas
df.loc[cantidad_filas]=["Venezuela",0,91164445]
df
```

```python
#Cambio de valor de una variable en una tupla
df.at[4,"Poblacion"]=[48108600]
df
```

```python
#Eliminacion por celda (valor de una variable en una tupla)
df2=df
df2.drop(1,axis=0)
```

**Ejercicio 04 -filtado de dataframes**

```python
#Operadoresde comparacion y logicos
#Que pais tiene mas de 46 millones
mas_de_46=df[df["Poblacion"]>48000000]
mas_de_46
```

```python
#Filtrado de operadores logico
df[(df["Poblacion"]>48000000)&(df["Superficie"]<6000000)]
```

```python
#Filtrado por valor de variable
por_nombre=df[df["Pais"]=="Colombia"]
por_nombre
```

```python
#filtrado por tamaño de valor de variable
nombre_largo=df[df["Pais"].str.len()>6]
nombre_largo
```

```python
#Filtrado por indice
por_nombre.index.to_list()[0]
```
