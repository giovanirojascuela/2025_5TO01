*Ejercicio 04 - filtrado de dataframe*

```python
#Operadoes de comparacion y logicos
#Que pais tiens mas de 46 millones
mas_de_46=df[df["Poblacion"]>48000000]
mas_de_46
```

```python
#Filtrando con operadores logicos
df[(df["Poblacion"]>48000000)&(df["Superficie"]<60000000)]
```

```python
#Filtrando por valor de variable
por_nombre=df[df["Pais"]=="Colombia"]
por_nombre
```

```python
#Filtrando por tamaño de variable
nombre_largo=df[df["Pais"].str.len()>6]
nombre_largo
```
