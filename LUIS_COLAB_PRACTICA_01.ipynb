**Ejercicio 03 - dataframe**

```python
import pandas as pd
# Crando un dataframe de un dataset
df=pd.DataFrame(data={"Pais":["Mexico","Argentina","España","Colombia"],
                      "Poblacion":[127212000,45167000,47099000,48922000]})
```

```python
# Mostrando Dataframe
df
```

```python
# Ordenar columnas por variable
df.sort_values(["Pais"])
```

```python
# Agregando columna o variable
df["Superficie"]=[1964375,2788400,505944,1142748]
df
```

```python
# Agregar solo un valor a las tuplas de una variable
df["Deporte"]="Futbol"
df
```

```python
# Eliminar columna o variable
df=df.drop(["Deporte"],axis=1)
df
```

```python
df2=df
df2
# Eliminacion multiples de vatiables
df2.drop(["Pais","Superficie"],axis=1)
```

```python
# Numero de tuplas
cantidad_filas=len(df)
cantidad_filas
```

```python
# Agrupacion de filas
df.loc[cantidad_filas]=["Venezuela",0,91164445]
df
```

```python
# Cambio de valor de una varuable en una tupla
df.at[4,"Poblacion"]=[48108600]
df
```

```python
# Eliminacion por celda(valor de una variable en una tupla)
df2=df
df2.drop(1,axis=0)
```

**Ejercicio 04 - filtrado de dataframes**

```python
# Operadores de comparacion y logicos
# Que pais tiene mas de 46 millones
mas_de_46=df[df["Poblacion"]>48000000]
mas_de_46
```

```python
# Filtrado con operadores logicos
df[(df["Poblacion"]>48000000)&(df["Superficie"]<6000000)]
```

```python
# Filtrado por valor de variable
por_nombre=df[df["Pais"].str.len()>6]
por_nombre
```

```python
# Filtracion por tamaño de variable
nombre_largo=df[df["Pais"].str.len()>6]
nombre_largo
```
