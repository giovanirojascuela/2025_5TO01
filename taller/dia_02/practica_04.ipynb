y 15, 16, 18, 19, 17, 16, 16, 17, 20, 16

```python
#importacion de lobrerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
#creando dataset
X = np.array([10, 11, 13, 15, 13, 12, 10, 11, 15, 10])
Y = np.array([15, 16, 18, 19, 17, 16, 16, 17, 20, 16])
print(X)
print(Y)
```

```python
# dividimos los datos en conjuntos de entrenamiento y pruebas
X_train, X_test, Y_train, Y_test=train_test_split(X.reshape(-1,1),
                                                  Y,test_size=0.2,
                                                  random_state=42)
print("entrenamiento X_______________________")
print(X_train)
print("prueba x_____________________________")
print(X_test)
print("entrenamiento Y_____________________")
print(Y_train)
print("prueba Y________________________")
print("Y_test")
```

```python
# creamos el modelo y lo entrenamremos
modelo=LinearRegression()
modelo.fit(X_train,Y_train)
```

```python
# Mostramos coeficientes
print("Coeficiente (pendiente):", modelo.coef_[0])
print("Intersección:", modelo.intercept_)
```

```python
# Graficamos los datos y la línea de regresión
# prediccion y grafico
plt.scatter(X,Y,color="blue",label="datos reales")
plt.plot(X,modelo.predict(X.reshape(-1,1)),color="red",label="linea de regresion")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
