#6.2 Comenzando manipulacion de dataframe




```python
import pandas as pd
```

```python
#creando un data set
df=pd.DataFrame(data={"Pais":["Mexico","Argentina","España","Colombia"],
                      "Poblacion":[127212000,45167000,47099000,48922000]})
```

```python
#mostramos el dataframe
df
```

```python
#ordenando columnas o variables
df.sort_values(["Pais"])
```

```python
#Agregar columna a variable
df["Superficie"]=[1964375,2788400,505944,1142748]
df
```

```python
#Agregar solo un valor unico
df["Deporte"]="futbol"
df
```

```python
#Eliminar columna
df=df.drop(["Deporte"],axis=1)
df
```

```python
#Eliminacion multiple de tablas
df.drop(["Pais","Superficie"],axis=1)
df
```

```python
#Agregar por filas
Cantidad_filas=len(df)
Cantidad_filas
```

```python
#Agregar filas
df.loc[Cantidad_filas]=["Benezuela",0,9116445]
df
```

```python
df.loc[4]=["Venezulea",0,9116445]
df
```

```python
#cambiar por celda
df.at[4,"Poblacion"]=[48108600]
```

```python
#eliminar por celda
df.drop(1,axis=0)
```

```python
# Queremos saber que pais tine mas de 46 millones
mas_de_46 = df[df["Poblacion"]>49600000]
mas_de_46
```

**6.3 Filtracion de dataframes**

```python
#Filtrando agregando operadores logicos
df[(df["Poblacion"]>46000000)&(df["Superficie"]<6000000)]
```

```python
por_nombre=df[df['Pais']=="Colombia"]
por_nombre
```

```python
#por tamaño
nombre_largos=df[df['Pais'].str.len()>6]
nombre_largos
```

```python
#por indice
por_nombre.index.to_list()[0]
```

```python
#Preparamos dataframe
#Agregamos columna con radio de habitantes por superficie
df["habitxkm2"]=(df["Poblacion"]/df["Superficie"]).astype(int)
df.sort_values(["habitxkm2"])
```
